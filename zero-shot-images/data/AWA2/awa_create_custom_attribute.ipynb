# label encoding

```python
cls_list = [ float(i) for i in range(1,51)]
```

```python
label_encoding = pd.DataFrame(cls_list).T
label_encoding.index +=1
```

```python
label_encoding.columns+=1
```

```python
label_encoding
```

```python
label_encoding.to_csv('label_encoding.csv',index=False)
```

# one hot encoding
https://stackoverflow.com/questions/52323575/generate-one-hot-encodings-from-dict-values

```python
cls_list = [ i for i in range(1,51)]
```

```python
# Sort and extract the indices.
idx = sorted(cls_list)
# Initialise a matrix of zeros.
aa = np.zeros((len(idx), max(idx)+1))
# Assign 1 to appropriate indices.
aa[np.arange(len(aa)), idx] = 1
```

```python
one_hot_encoding = pd.DataFrame(aa)
one_hot_encoding = one_hot_encoding.drop(0, 1)
one_hot_encoding.index +=1
one_hot_encoding
```

```python
one_hot_encoding.to_csv('one_hot_encoding.csv',index=False)
```

# bit encoding
https://stackoverflow.com/questions/49836933/encode-integer-pandas-dataframe-column-to-padded-16-bit-binary

```python
df = pd.DataFrame(dict(A=range(50)))
```

```python
a = df.A.values
n = int(np.log2(a.max() + 1))
b = (a[:, None] >> np.arange(n)[::-1]) & 1
c = b.astype(np.float32)
bit_encoding = pd.DataFrame(c).T
bit_encoding.index +=1
bit_encoding.columns +=1
bit_encoding
```

```python
bit_encoding.to_csv('bit_encoding.csv',index=False)
```
